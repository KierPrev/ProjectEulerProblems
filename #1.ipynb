```julia
i = 1
resultado = 0

multiplos = Set{Int}()
```

#### Multiplos de 3

```julia
while 3*i < 1000    
    resultado = 3*i
    push!(multiplos, resultado)
    i += 1
end

println(multiplos)
i = 1
```

#### Multiplos de 5

```julia
while 5*i < 1000    
    resultado = 5*i
    push!(multiplos, resultado)
    i += 1
end

println(multiplos)
```

#### Suma de todos los multiplos en set

```julia
sum(multiplos)
```
